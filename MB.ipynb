## Q3. $\pi$ from Monte-Carlo



```python
import numpy as np
import matplotlib.pyplot as plt
```

## Using `random_bisection()`

```python
def random_bisection(a,b):
  tol = 1e-5
  diff = abs(b - a)
  while diff > tol:
    choice = np.random.choice([a,b])

    if choice == a:
        b = (a + b) / 2
    else:
        a = (a + b) / 2
    diff = abs(b - a)


  return (a+b)/2
```

```python
def r(x,y):
 z = np.sqrt(x**2 + y**2)
 return z
```

```python
epoch_values = np.arange(1, 101, 1)
estimated_pi_values = []

for num_samples in epoch_values:
    count_within_circle = 0
    x_values = [random_bisection(0, 1) for _ in range(num_samples)]
    y_values = [random_bisection(0, 1) for _ in range(num_samples)]


    for j in range(num_samples):
        z = r(x_values[j], y_values[j])
        if z <= 1:
            count_within_circle += 1

    estimated_pi = 4 * (count_within_circle / num_samples)
    estimated_pi_values.append(estimated_pi)

print(f"Epoch values: {epoch_values}")
print(f"Estimated Pi values: {estimated_pi_values}")
```

```python
plt.figure(figsize=(10, 6))
plt.plot(epoch_values, estimated_pi_values, label='Estimated Pi')

tensor_same_shape = np.full_like(epoch_values, np.pi, dtype=float)
pi_values = np.array(tensor_same_shape)
print(np.mean(estimated_pi_values))

plt.plot(epoch_values, pi_values, label='Actual Pi', linestyle='--')
plt.title('Estimated Pi vs. Actual Pi')
plt.xlabel('Number of Epochs')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
error = np.abs(estimated_pi_values - pi_values)/pi_values
print(error)
```

```python
#plt Error
plt.plot(epoch_values, error, label='Estimated Pi')
plt.title('Error')
plt.xlabel('Number of Epochs')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.savefig('error.png')
plt.show()
```

# Q4. Sampling from different distribution

$$P(x) = exp(-x) $$
$$ x \, \epsilon \,\,[0,∞) $$

CDF:  $\,\,\,\,\,\,\,\,\, F(x) = - \int_{0}^{x}P(x')dx'= \int_{0}^{x} exp(-x')dx' $

$\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,F(x) = 1 - exp(-x)$

Inverse: $ F^{-1}(x) = -ln(1-F) $

as $x \, \epsilon \,\,[0,∞) \implies P(x) \, \epsilon  \, (0,1) \implies F(x) \, \epsilon \, (0,1)$

so $F^{-1}(x)\, \epsilon \, (0,∞)$



```python
def random_bisection(a,b):
    tol = 1e-5
    diff = abs(b - a)
    while diff > tol:
        choice = np.random.choice([a,b])

        if choice == a:
            b = (a + b) / 2
        else:
            a = (a + b) / 2
        diff = abs(b - a)


    return choice
```
